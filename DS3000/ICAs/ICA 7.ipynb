A logarithmic birthday party is the moment you are some integer power of 10 seconds old (1 second old, 10 seconds old, 100 seconds old, ...)

Compute your first 10 logarithmic birthday parties, store them in a dataframe as shown below.  (You're welcome to use a fake birthday if you'd like)

|  log_bday |  day | hour | minute | month |  sec |   year |
|----------:|-----:|-----:|-------:|------:|-----:|-------:|
|  10^0 sec |  1.0 |  7.0 |   25.0 |   1.0 |  1.0 | 1985.0 |
|  10^1 sec |  1.0 |  7.0 |   25.0 |   1.0 | 10.0 | 1985.0 |
|  10^2 sec |  1.0 |  7.0 |   26.0 |   1.0 | 40.0 | 1985.0 |
|  10^3 sec |  1.0 |  7.0 |   41.0 |   1.0 | 40.0 | 1985.0 |
|  10^4 sec |  1.0 | 10.0 |   11.0 |   1.0 | 40.0 | 1985.0 |
|  10^5 sec |  2.0 | 11.0 |   11.0 |   1.0 | 40.0 | 1985.0 |
|  10^6 sec | 12.0 | 21.0 |   11.0 |   1.0 | 40.0 | 1985.0 |
|  10^7 sec | 27.0 |  1.0 |   11.0 |   4.0 | 40.0 | 1985.0 |
|  10^8 sec |  3.0 | 17.0 |   11.0 |   3.0 | 40.0 | 1988.0 |
|  10^9 sec |  9.0 |  9.0 |   11.0 |   9.0 | 40.0 | 2016.0 |

(++) add a column with a more easily readable time (e.g. `February-11 Feb:02`)

```python
from datetime import datetime, timedelta
import pandas as pd

df_log_bday = pd.DataFrame()

# my birthday
bday = datetime(year=2001, month=12, day=2, hour=8, minute=20, second=0)

for power in range(11):
    # to offset each log bday
    offset = timedelta(seconds=(10 ** power))
    
    # new birthday after each log bday
    bday = bday + offset
    
    # outputting the table
    bday_series = pd.Series({'log_bday': f'10{power} sec', 
                 'day': bday.day, 
                 'hour': bday.hour, 
                 'minute': bday.minute, 
                 'month': bday.month, 
                 'sec': bday.second, 
                 'year': bday.year})
    
    df_log_bday = df_log_bday.append(bday_series, ignore_index=True)
    
df_log_bday
```
