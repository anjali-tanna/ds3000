## In Class Activity 1

Osaka, Japan is located at:

    34.6937° N, 135.5023° E
    
1. Create a dataframe of the next 24 hours of their weather as was done above
2. Make a function `get_forecast` which accepts:
    - `lat`
    - `lon`
    - `api_key`
    - `units` (default = 'imperial')
    
    and returns a dataframe of the next 24 hours of the location's weather.

````python
import requests
import json
import pandas as pd
from datetime import datetime

# input arguments to API (these define our query)
# north = positive, south = negative
lat = 34.6937

# west = positive, east = negative
lon = -135.5023

# api key
api_key = 'a09fbc60af08d36a8942789ae2864dab'


def get_forecast(lat, lon, 
                 api_key='a09fbc60af08d36a8942789ae2864dab', 
                 units='imperial'):
    
    # build url
    url = f'https://api.openweathermap.org/data/2.5/onecall?lat={lat}&lon={lon}&appid={api_key}&units={units}'
    print(url)

    # get url response as a string
    url_text = requests.get(url).text

    # convert from json to a nested dict
    weather_dict = json.loads(url_text)

    # build dataframe from nested dict
    df_daily = pd.DataFrame()
    for day_dict in weather_dict['daily']:
        
        # we add a 'datetime' feature from the unix time (easier to use)
        day_dict['datetime'] = datetime.fromtimestamp(day_dict['dt'])

        # build a single row of dataframe (weather prediction for some hour)
        s_day = pd.Series(day_dict)

        # collect forecasts in a dataframe
        df_daily = df_daily.append(s_day, ignore_index=True)

    return df_daily.head()

get_forecast(lat, lon)
````

## In Class Activity 2
1. Write a function `from_unix_to_datetime()` which:
- accepts:
    - unix_time (float): time since jan 1 1970 (in computer's time zone)
    - timezone_from (str): timezone of input unix_time ('uct' by default)
    - timezone_to (str): timezone of output datetime object
- returns:
    - datetime_tz (datetime): datetime object in given timezone
Be sure to properly document your function.
2.  Write a few `assert` test cases to validate your function.
    - Rather than checking if values are proper yourself, it may be best to rely on [another source](https://www.epochconverter.com/)

````python
import requests
import json
from datetime import datetime, date, time
import pytz

def from_unix_to_datetime(unix_time, timezone_to='US/Eastern'):
    """ converts unix time to datetime object
    
    Args: 
        unix_time (float): unix time (sec since jan 1 1970 UTC)
        timezone_to (str): timezone of output datetime object
        
    Returns:
        datetime_tz (datetime): datetime object in given timezone
    """
    
    # build datetime
    datetime_from = datetime.utcfromtimestamp(unix_time)
    
    # add timezone to datetime
    tz_from = pytz.timezone('UTC')
    datetime_from = tz_from.localize(datetime_from)
    
    # convert timezone
    tz_to = pytz.timezone(timezone_to)
    dt = datetime_from.astimezone(tz_to)
    
    return dt
````

````python
# test case

unix_time = 1644940967
timezone_to = 'US/Eastern'

from_unix_to_datetime(unix_time=1644940967, timezone_to='US/Eastern')

datetime_computed = from_unix_to_datetime(unix_time=1644940967, timezone_to='US/Eastern')

time_zone = pytz.timezone(timezone_to)

datetime_expected = time_zone.localize(datetime(2022, 2, 15, 11, 2, 47))

assert datetime_expected == datetime_computed
````
