```python
import seaborn as sns

# may take a 15 sec on first run to download titanic data
df_titanic = sns.load_dataset('titanic')

s_titanic_fare = df_titanic['fare']

s_titanic_fare
```

```python
df_titanic
```

```python
s_titanic_fare.describe()

# the count is the amount of people who paid to get on the titanic (891 people)
# the mean is the average price a person paid ($32)
# the std is how much each price paid deviates from the mean
# the min is the minimum amount someone paid ($0)
# the 25% is the 25th percentile (25% paid $7.9 or less)
# the 50% is the median
# the 75% is the 75th percentile (75% paid $31 or less)
# the max is the maximum amount someone paid ($512)
```
